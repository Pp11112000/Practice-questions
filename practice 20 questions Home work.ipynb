3 . Write a funcation which will take input as a dict and give me out as a list of all the values 
even in case of 2 level nesting it should work .

```python
def get_values(d):
    for v in d.values():
        if isinstance(v, dict):
            yield from get_values(v)
        else:
            yield v


a = {'4': 'hello', '6': 'hi', '7': {'8': 'happy', '9': 'new', '5': {'10': 'year'}, '2': '2022', '6': {'2': 7, '1': 8}}}

print(list(get_values(a)))
```

9 . write a function which will be able to show date and time  

```python
#Here, we imported the datetime class from the datetime module
from datetime import datetime
def date_time():
    #Then, we used the datetime.now() method to get the current local date and time
    now = datetime.now()
    #print(now)
    dt_string = now.strftime("Date: "'%B-%d-%Y' '\n'"Time: "'%H:%M:%S')
    print(dt_string)
```

```python
date_time()
```
